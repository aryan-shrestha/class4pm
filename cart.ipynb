```python
#WAP for result management system

no_of_student = int(input("Enter number of student"))
students = []

for i in range(no_of_student):
    name = input("Enter name of student: ")
    physics = int(input("Enter marks in physics"))
    chemistry = int(input("Enter marks in chemistry"))
    biology = int(input("Enter marks in biology"))
    nepali = int(input("Enter marks in nepali"))
    math = int(input("Enter marks in math"))
    total = physics + chemistry + biology + nepali + math
    percentage = total/0.5
    result = {'name':name, 'physics': physics, 'chemistry': chemistry, 'biology': biology, 'nepali': nepali, 'math': math,
             'total': total, 'percentage': percentage}
    students.append(result)
    
print("/n Results :")

for student in students:
    print(student)

def update_result(student_name):
    for student in students:
        if student.get('name') == student_name:
            student['physics']= int(input("Enter new marks in physics"))
            student['chemistry']= int(input("Enter new marks in chemistry"))
            student['biology']= int(input("Enter new marks in biology"))
            student['nepali']= int(input("Enter new marks in nepali"))
            student['math']= int(input("Enter new marks in math"))
            student['total']= student['physics'] + student['chemistry'] + student['biology']+ student['nepali'] + student['math']
            student['percentange'] = student['total']*0.2
            user_input()
            
    
def user_input():
    user_decison = input("\nDo you want to edit the result(yes/no): ")
    if user_decison == 'yes':
        student_name = input("\nEnter the name of student you want to change: ")
        update_result(student_name)
    else:
        print('\n Updated Results: \n')
        for student in students:
            print(student)
user_input()
```
